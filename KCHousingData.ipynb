```python
import pandas as pd

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

import numpy as np


from dmba import stepwise_selection
from dmba import AIC_score
```

```python
df = pd.read_csv("data/kc_house_data.csv")
```

```python
# In multiple linear regression, each new predictor has a coefficient and they're all added
# together, SIMPLY
# Each predictor contributes independently and proportionally to the target, as though others were not there
# Meaning you could take each predictor and the target variable and create a simple linear model with it
```

```python
# MODEL SELECTION 

# In multiple linear regression, when choosing which model is the best, using RMSE or R2 is not usually the best
# Adding more variables naturally increases R2 and reduces RMSE for the training data

# Use AIC, AICc, BIC or Mallows Cp, R2 adjusted
# AIC penalizes adding terms; k variables penalized by 2k
# So our goal is to minimize AIC or maximize R2 adjusted

# subset regression: trying every possible model
# stepwise
  # - forward selection
  # - backward elimination
  # - adding and removing while comparing AIC and adjusted R2
```

```python
# We start with a few predictors
predictors = [
    "sqft_living", 
    "sqft_lot", 
    "bedrooms", 
    "bathrooms", 
    "grade"
]
outcome = "price"

x, y = df[predictors], df[outcome]
```

```python
sk_model = LinearRegression()
sk_model.fit(x, y)
```

```python
fitted_base = sk_model.predict(x)

r2_base = r2_score(y, fitted_base)
rmse_base = np.sqrt(mean_squared_error(y, fitted_base))

print(f"R2: {r2_base:.2f}")
print(f"RMSE: {rmse_base:.2f}")
```

```python
print(f'Intercept: {sk_model.intercept_:.3f}')
print('Coefficients:')

for name, coef in zip(predictors, sk_model.coef_):
    print(f' {name}: {coef}')
```

```python
# Let's use statsmodels to check things out
sm_model = sm.OLS(y, x.assign(const=1))
results = sm_model.fit()

results.summary()
```

```python
# Adding more predictor variables
predictors_2 = predictors + ["sqft_basement", "waterfront", "yr_built", "yr_renovated"]
x2 = df[predictors_2]
```

```python
x2["sqft_basement"] = pd.to_numeric(x2["sqft_basement"], errors="coerce")
```

```python
sm_model = sm.OLS(y, x2.assign(const=1))
results = sm_model.fit()

results.summary()

# Adjusted R² is always ≤ R²
# The new variables (sqft_basement, waterfront, yr_built, yr_renovated) are actually adding real predictive value, not just noise.
# If:

# your dataset is large
# and your added variables are useful

# 👉 then the penalty term becomes tiny

# So:

# Adjusted R² ≈ R²
# Difference becomes almost invisible

# A significant difference between R2 and Adjusted R2 is a sign of overfitting
```

```python
print("r2: ", results.rsquared)
print("r2 adj: ", results.rsquared_adj)

# r2 adj is less than r2 as expected. Although not significantly, which is a good thing

# Also, AIC is significantly different between the two models: 427,500−423,600=3,900
# <2: no real difference, 4 to 10: some, >10 strong, >>10 very strong evidence
# You can only compare AIC values when:
#   - Models are fit on the same dataset
#   - Same target variable
```

```python
results.rsquared - results.rsquared_adj
```

```python
# removing sqft_basement and yr_renovated since their p values showed they're insignificant in the second model
predictors_3 = predictors + ["waterfront", "yr_built"]
x3 = df[predictors_3]

sm_model_3 = sm.OLS(y, x3.assign(const=1))
results = sm_model_3.fit()

results.summary()

# turns out they were truly insignificant since r-squared adj and AIC remained thesame
```

```python
# Stepwise Regression
# stepwise_selection, forward_selection, and backward_elimination

X = x2
def train_model(variables):
    if len(variables) == 0:
        return None
    model = LinearRegression()
    model.fit(X[variables], y)
    return model


def score_model(model, variables):
    if len(variables) == 0:
        return AIC_score(y, [y.mean()] * len(y), model, df=1)
    return AIC_score(y, model.predict(X[variables]), model)
    
```

```python
best_model, best_variables = stepwise_selection(
    X.columns, 
    train_model, 
    score_model, 
    verbose=True
)

print()
print(f'Intercept: {best_model.intercept_:.3f}')
print('Coefficients:')
for name, coef in zip(best_variables, best_model.coef_):
    print(f' {name}: {coef}')
```
